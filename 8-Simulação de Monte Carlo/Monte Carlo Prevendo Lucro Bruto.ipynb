Requisitos:

-Receita(revenue/rev) Esperada

-CMV/COGS esperado ( Custo da Mercadoria Vendida)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m= 170 # receita média
rev_stdev= 20 # desvio padrão da receita
iterations= 1000 # Nº de simulaçoes
```

```python
rev= np.random.normal(rev_m, rev_stdev, iterations)
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
COGS= - (rev * np.random.normal(0.6, 0.1)) #  0.6, 0.1: Média e Desvio padrão, respectivamente# Atenção ao sinal de menos nessa equaçao, pois se trata de gasto de dinheiro

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

Parte II

Calculando o Lucro Bruto

```python
Gross_Profit= rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```
